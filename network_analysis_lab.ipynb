## 3- Counting Triangles in Graphs

```python

def compute_triangles(Graph):
    dict_v = {}
    for v in Graph.nodes():
        dv = 0
        for a in nx.neighbors(Graph, v):
            for b in nx.neighbors(Graph, v):
                if a!=b and a in nx.neighbors(Graph, b): # undirected graph propreties
                    dv+=1
        # because we counted the pair (a,b) and (b,a)
        dict_v[v] = dv//2
    
    return dict_v
                    
    
```

```python
print(compute_triangles(G))
print(nx.triangles(G))
```

```python
sum(compute_triangles(G).values())/3
```

```python
# combination calculation function
def combinaison(n,r):
    f = math.factorial
    return f(n) // f(r) // f(n-r)
```

```python
combinaison(5,3)
```

## 4-community detection

```python
g_tmp = G.copy()
pos = nx.kamada_kawai_layout(g_tmp)
nx.draw_networkx(g_tmp,pos, with_labels=True)
plt.show()
```

```python
#nx.is_empty(G)

number_of_communities = []

modularity_of_community = []

communities = []

max_modularity_index = -1
max_modularity_value = -1

while not nx.is_empty(g_tmp):
    edge_centrality = nx.edge_betweenness_centrality(g_tmp, k=None, normalized=True, weight=None, seed=None)
    edge = max(edge_centrality, key=edge_centrality.get)
    g_tmp.remove_edge(edge[0],edge[1])
    
    com = list(nx.connected_components(g_tmp))
    
    if not len(com) in number_of_communities: # removing the edge didnt create new communities
        number_of_communities.append(len(com))
        communities.append(com)
        value = nx.algorithms.community.modularity(G, list(nx.connected_components(g_tmp)))
        modularity_of_community.append(value)
        
        #test if it is the max
        if value>max_modularity_value:
            max_modularity_value = value # max value
            max_modularity_index = len(modularity_of_community) -1 # index of last element added
    
    print((list(nx.connected_components(g_tmp))))
    print(nx.algorithms.community.modularity(G, list(nx.connected_components(g_tmp))))
        

#pos = nx.kamada_kawai_layout(g_tmp)
#nx.draw_networkx(g_tmp,pos, with_labels=True)
#plt.show()

    
```

```python

plt.xlabel('number of communities')
plt.ylabel('modularity rate')
plt.title('madularity in relation with number of communities ');
plt.plot(number_of_communities,modularity_of_community)
```

```python
print(" max value of modularity is %f"%max_modularity_value)
print(" number of correspending communities is %d "%number_of_communities[max_modularity_index])
```

```python
def get_community_number(communities,node):
    # returns the index of the community in which node belongs
    index = -1
    for community in communities:
        index +=1
        if node in community:
            break

    return index
```

```python
available_colors = ["#95a5a6","#f39c12","#2ecc71","#9b59b6","#2980b9","#013243"]

print(" communities are :")
print(communities[max_modularity_index])

color = []
i=0
for n in G.nodes():
    color.append(available_colors[get_community_number(communities[max_modularity_index],n)])

pos = nx.kamada_kawai_layout(G)
#nx.draw_networkx(G, pos, node_color=communities[max_modularity_index])
nx.draw_networkx(G,pos, node_color=color, with_labels=True)
plt.show()
```
